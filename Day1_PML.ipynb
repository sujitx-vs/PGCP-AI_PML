Dataframe- 2D combination of rows & Columns

```python
student ={
    "Name":["Amit","Bob","Tom"],
    "Marks":[78,56,89]
}

df= pd.DataFrame(student)
```

```python
df
```

```python
df["Grade"]=["A","B","C"]
```

```python
df
```

```python
df = df.drop("Grade",axis = 1)
```

```python
df
```

```python
df.rename(columns = {"Name":'Title'},inplace=True)
df
```

```python
df["Marks"]= df["Marks"].apply(lambda x:x+5)
df
```

```python
df
new_data = pd.DataFrame({
    'Title': ['Ram', 'Raju','Kesu'],
    'Marks': [82, 78,93]

})

df = pd.concat([df, new_data], ignore_index=True)
```

```python
df
```

```python
# lambda function to find the marks greater than 90 or less than 76.
```

```python
[c for c in df["Marks"] if c >90]
```

```python
filtered_df = df[(df['Marks'] > 90) | (df['Marks'] < 76)]
```

```python
filtered_df
```
