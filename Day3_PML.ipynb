```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = {
    'Income': [50000, 60000, 55000, np.nan, 65000],
    'Gender': ['Male', 'Female', 'Female', 'Male', np.nan],
    'City': ['Mumbai', 'Delhi', 'Mumbai', 'Chennai', 'Delhi'],
    'Signup_Date': ['2022-01-01', '2021-06-15', '2022-03-10', '2020-12-20', '2021-08-05'],
    'Last_Login': ['2023-01-01', '2023-02-10', '2023-01-15', '2023-01-20', '2023-02-01'],
    'Purchased': [1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['Signup_Date'] = pd.to_datetime(df['Signup_Date'])
df['Last_Login'] = pd.to_datetime(df['Last_Login'])
```

```python
df.info()
```

```python
# 1. Handling Missing Values.
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

#SimpleImputer.

#Numerical column

imputer = SimpleImputer(strategy = "mean")
df['Income'] = imputer.fit_transform(df[['Income']])
```

```python
df
```

```python
#Categorical Column
imputer_cat = SimpleImputer(strategy = "most_frequent")
df['Gender'] = imputer_cat.fit_transform(df[['Gender']]).ravel()
df
```

```python
#Data Feature Engineering.

df['Signup_Year'] = df['Signup_Date'].dt.year
df['Signup_Month'] = df['Signup_Date'].dt.month
```

```python
df
```

```python
df["Tenure_Days"] = (df['Last_Login']-df['Signup_Date']).dt.days
```

```python
df
```

```python
#Onehot encoding

encoder = OneHotEncoder()
city = encoder.fit_transform(df[['City']])

city_df = pd.DataFrame(city.toarray(),columns=encoder.get_feature_names_out(['City']))
df = pd.concat([df,city_df],axis = 1)
```

```python
df
```

```python
#feature Scaling

scalar = StandardScaler()
df[['Income_1']] = scalar.fit_transform(df[['Income']])
```

```python
df
```

```python
#Feature Transform
df['Income_log'] = np.log1p(df['Income'])
```

```python
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.model_selection import  train_test_split
```

```python
data = {
    'Income': [50000, 60000, 55000, np.nan, 65000],
    'Gender': ['Male', 'Female', 'Female', 'Male', np.nan],
    'City': ['Mumbai', 'Delhi', 'Mumbai', 'Chennai', 'Delhi'],
    'Signup_Date': ['2022-01-01', '2021-06-15', '2022-03-10', '2020-12-20', '2021-08-05'],
    'Last_Login': ['2023-01-01', '2023-02-10', '2023-01-15', '2023-01-20', '2023-02-01'],
    'Purchased': [1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)
df
```

```python
num_col = ['Income']
cat_col = ["Gender","City"]
```

```python
from sklearn import pipeline
n_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="mean")),
    ('scalar',StandardScaler())
])

c_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="most_frequent")),
    ('encoder',OneHotEncoder(handle_unknown='ignore'))
])


preprocessor = ColumnTransformer(
    [
        ('num',n_pipeline,num_col),
        ("cat",c_pipeline)
    ]
)


pipeline = Pipeline([
    ('preprocessor',preprocessor),
    ('model', LogisticRegression())
])
```
