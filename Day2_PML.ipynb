Handling of missing Values

  - missing values are data points that are not recorded or remain absent in a dataset.
  - They are generaly represented as NAN or Blank spaces.

Technique to Handle Missing Values

A. Removing missing Values:
  - Very small amount of missing data.
  - Column is not immportant.

Adv :
  - Simple and fast
  - No data disortion

Disadv:
  - Data loss
  - Not suitable for small dataset.

B. mean Imputation
  - Replace missing values with the average values

usage :     
  - numeric data
  - Not suitable for existing outliers.

adv ;
  - Easy to implement.
  - Preserve the size of datasets.

Disadv ;
  - Redduce variance
  - Sensitive to Outliers.


C. Median Imputations



D. Mode Imputations:
  - Replace missing values with the most frequent values.

Usage;
  - Categorical data (city,gender)

adv-
  - Best for categorical data
  - simple approach
Disadv-
  - can create model bias towards frequent values.

NB: if the data is left or right skwed then median is the best choice for replacement.




E. Forward Fill(ffill)

### fill backword

  - FIll missing values using the next value

  - Usage: Forecasting.

  - Adv:
    - Useful when future data is reliable.
  - Disadv:
    - It will introduce unrealistic backward assumptions.


G. Interpolation
  - Estimates distinct missing values using mathematical patterns between data points.

- Usage
  - time series and continous data.

- Adv - Give more distinct values than mean/median

- Disadv - May fall with irregular data

```python
data = {
    "Time": ["T1", "T2", "T3", "T4"],
    "Temp": [30, np.nan, np.nan, 35]
}

df = pd.DataFrame(data)
df
```

```python
df['Temp'].fillna(method='bfill',inplace= True)
df
```

```python
data = {
    "Time": ["T1", "T2", "T3", "T4"],
    "Temp": [30, np.nan, np.nan, 35]
}

df = pd.DataFrame(data)
df
```

```python
df["Temp"] = df["Temp"].interpolate()
df
```

H. Constant Value imputation:
  - Replace the missing values with a fixed value like unknown or 0.

adv
- simple
- Useful for categorical labeling.

disadv:
can disort the data distibution.


```python
data = {
  "Name": ["A", np.nan, "C", "D"],
  "Age": [25, np.nan, 30, np.nan],
  "Marks": [80, 90, np.nan, 85]
}

df = pd.DataFrame(data)
df
```

```python
df["Marks"].fillna(0,inplace=True)
df["Age"].fillna(22,inplace=True)
df["Name"].fillna("Unknown",inplace=True)
df
```
